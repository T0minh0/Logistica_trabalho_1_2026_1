# 03. Previsão de Demanda

Este notebook treina e avalia modelos de previsão (Baselines, ETS, ARIMA) nos dados de séries temporais.

```python
import sys
sys.path.append('..')
from src.io_load import load_m5_data
from src.preprocess import preprocess_data
from src.forecast.ets_arima import ETSForecaster
from src.forecast.metrics import evaluate_forecast

# Configurar Dados
sales_raw, calendar, prices = load_m5_data()
df = preprocess_data(sales_raw, calendar, prices)

sample_item = df['item_id'].unique()[0]
sample_store = df['store_id'].unique()[0]
subset = df[(df['item_id'] == sample_item) & (df['store_id'] == sample_store)].sort_values('date')

# Divisão
train = subset.iloc[:-28]
test = subset.iloc[-28:]

y_train = train['demand']
y_test = test['demand']
```

```python
# Treinar ETS
ets = ETSForecaster()
ets.fit(y_train)
y_pred = ets.predict(28)

# Avaliar
metrics = evaluate_forecast(y_test, y_pred)
print(metrics)
```
